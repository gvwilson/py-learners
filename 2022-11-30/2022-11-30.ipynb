# Statistical Tests

```python
import pandas as pd
from scipy.stats import ttest_ind # independent two-sample t-test
```

```python
data_1 = pd.read_csv("luminance-001.csv")
data_1["source"] = "1"
data_1
```

```python
data_2 = pd.read_csv("luminance-002.csv")
data_2["source"] = "2"
data_2
```

```python
# Compare everything to everything.
ttest_ind(data_1["reading"], data_2["reading"])
```

```python
data_1["row_id"].isin(["B", "C"])
```

```python
data_1["row_id"].isin(["B", "C"]) & data_1["col_id"].isin([2, 3])
```

```python
def center(df):
    return df[df["row_id"].isin(["B", "C"]) & df["col_id"].isin([2, 3])]

center(data_1)
```

```python
center(data_2)
```

```python
ttest_ind(center(data_1)["reading"], center(data_2)["reading"])
```

```python
def boundary(df):
    return df[~ (df["row_id"].isin(["B", "C"]) & df["col_id"].isin([2, 3]))]

boundary(data_1)
```

```python
# Center versus boundary
ttest_ind(center(data_1)["reading"], boundary(data_1)["reading"])
```

```python
# What if we *don't* assume equal variances?
ttest_ind(center(data_1)["reading"], boundary(data_1)["reading"], equal_var=False)
```

```python
# Paired samples: use ttest_rel instead of ttest_ind
from scipy.stats import ttest_ind #
ttest_rel(center(data_1)["reading"], center(data_2)["reading"])
```
