```python
def read_data(pattern):
    filenames = glob.glob(pattern)
    dataframes = []
    for current in filenames:
        temp = pd.read_csv(current)
        temp["filename"] = current
        dataframes.append(temp)
    result = pd.concat(dataframes).reset_index()
    return result
```
